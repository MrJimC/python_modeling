### Build up a dataset need to be fit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1, 1, 0.02)
y = 2 * np.sin(x * 2.3) + np.random.rand(len(x))
```

```python
plt.scatter(x, y)
plt.show()
```

### Using `np.polyfit` to fit a line with degree of 3

```python
parameter = np.polyfit(x, y, 3)
```

```python
# 输出的结果为3次方程的参数,我们可以像下面这样把方程拼接出来
y2 = parameter[0] * x ** 3 + parameter[1] * x ** 2 + parameter[2] * x + parameter[3]
```

```python
# print the fitting result
plt.scatter(x, y)
plt.plot(x, y2, color='g')
plt.show()
```

```python
# We can also use poly1d() to  拼接方程式

p = np.poly1d(parameter)
plt.scatter(x, y)
plt.plot(x, p(x), color='g')
plt.show()
```
